```python
# One additional option we have for plotting one quantitative variable against one qualitative variable is
# the adaptation of bar charts
# Here, instead of showing the counts on the vertical axis, we can show the mean of a second variable
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv('./fuel_econ.csv')
fuel_econ.shape
```

```python
fuel_econ.head()
```

```python
# charting bax plots
base_color = sb.color_palette()[0]
sb.boxplot(data=fuel_econ, x='VClass', y='comb', color=base_color);
plt.xticks(rotation=15);
```

```python
# charting a bar plot for the above
base_color = sb.color_palette()[0]
sb.barplot(data=fuel_econ, x='VClass', y='comb', color=base_color);
plt.xticks(rotation=15);
plt.ylabel('Avg. Combined Fuel Eff. (mpg)');
```

```python
# The black tick marks on each bar indicate uncertainty about the mean as a confidence info
# If we want to leave the black ticks out of the graph - use the code below
base_color = sb.color_palette()[0]
sb.barplot(data=fuel_econ, x='VClass', y='comb', color=base_color, errwidth=0);
plt.xticks(rotation=15);
plt.ylabel('Avg. Combined Fuel Eff. (mpg)');
```

```python
# Alternatively, we could change the bars to represent the standard deviation of data points by using the below code
base_color = sb.color_palette()[0]
sb.barplot(data=fuel_econ, x='VClass', y='comb', color=base_color, ci='sd');
plt.xticks(rotation=15);
plt.ylabel('Avg. Combined Fuel Eff. (mpg)');
```

```python
# We could also use a point plot
sb.pointplot(data=fuel_econ, x='VClass', y='comb', ci='sd');
plt.xticks(rotation=15);
plt.ylabel('Avg. Combined Fuel Eff. (mpg)');
```

```python
# we can get rid of connecting line if we would like to with the below code
sb.pointplot(data=fuel_econ, x='VClass', y='comb', ci='sd', linestyles="");
plt.xticks(rotation=15);
plt.ylabel('Avg. Combined Fuel Eff. (mpg)');
```
